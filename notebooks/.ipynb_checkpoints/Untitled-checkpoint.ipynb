# Single-Path Impulse Geometry

This notebook implements the single-path impulse-driven trajectory
described in Section X of the Impulse Geometry foundational paper.

We construct a single evolving state driven by a local impulse field,
integrate it forward, and examine the resulting curve.

    

```python
import numpy as np
import matplotlib.pyplot as plt

from iglab.core.state import State
from iglab.systems.single_path import SinglePathSystem
```

```python
def impulse(state):
    """
    Constant impulse: straight-line motion.
    """
    return np.array([1.0, 0.0])
```

```python
s0 = State(
    x=np.array([0.0, 0.0]),
    s=0.0
)
```

```python
sys = SinglePathSystem(
    initial_state=s0,
    impulse_fn=impulse,
    dt=0.05
)

sys.run(200)
traj = np.array(sys.trajectory())
```

```python
plt.figure(figsize=(5, 5))
plt.plot(traj[:, 0], traj[:, 1])
plt.axis("equal")
plt.title("Single-Path Trajectory (Constant Impulse)")
plt.show()
```

```python
def curved_impulse(state):
    return np.array([
        np.cos(state.s),
        np.sin(state.s)
    ])
```

```python
sys = SinglePathSystem(
    initial_state=s0,
    impulse_fn=curved_impulse,
    dt=0.05
)

sys.run(400)
traj = np.array(sys.trajectory())

plt.figure(figsize=(5, 5))
plt.plot(traj[:, 0], traj[:, 1])
plt.axis("equal")
plt.title("Impulse-Driven Curved Path")
plt.show()
```
