# IG Fractal Tree (Theory-Accurate)

Impulse space: U(s) = (ρ(s), κ(s)) ∈ R²
State: X(s) = (x(s), y(s), θ(s)) ∈ R³

Generator vector field:
x'(s) = ρ(s) cos θ(s)
y'(s) = ρ(s) sin θ(s)
θ'(s) = κ(s)

Branching event at s = s_b:
- State inheritance: X_child(0) = X_parent(s_b)
- Impulse inheritance (binary): ρ_child(s) = ρ_parent(s), κ_child(s) = ± κ_parent(s)
- No geometric angle offsets at the branch point.


```python
import copy
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 0.72              # 0 < A < 1
k0 = 0.2              # turning-rate scale

def rho(s):
    # exponential amplitude scaling ρ(s) = A^s
    return A**s

def kappa(s):
    # analytic κ(s): constant is analytic and simplest
    return k0
```

```python
class State:
    def __init__(self, x, y, theta, s):
        self.x = float(x)
        self.y = float(y)
        self.theta = float(theta)
        self.s = float(s)

def euler_step(st, dt, kappa_sign=+1):
    # Evaluate impulse at current s (global parameter)
    r = rho(st.s)
    k = kappa_sign * kappa(st.s)

    # ODE: x' = ρ cosθ, y' = ρ sinθ, θ' = κ
    st.x += dt * r * np.cos(st.theta)
    st.y += dt * r * np.sin(st.theta)
    st.theta += dt * k
    st.s += dt
    return st
```

```python
def spawn_children(parent_end_state):
    """
    Binary branching:
      left:  κ -> +κ
      right: κ -> -κ
    State is inherited exactly at branch point.
    """
    child0 = copy.deepcopy(parent_end_state)
    child1 = copy.deepcopy(parent_end_state)
    return [(child0, +1), (child1, -1)]
```

```python
def generate_tree(depth_max=9, dt=0.005):
    segments = []

    # root: first branchpoint; two symmetric branches are typical (paper mentions this)
    # We'll start at origin with upward orientation.
    root = State(x=0.0, y=0.0, theta=np.pi/2, s=0.0)

    # queue items: (state_at_branchpoint, depth_n, kappa_sign)
    queue = [(root, 0, +1), (root, 0, -1)]

    while queue:
        st0, depth_n, sign = queue.pop()

        pts, st_end = integrate_generation_segment(st0, depth_n, dt=dt, kappa_sign=sign)
        segments.append(pts)

        if depth_n + 1 <= depth_max:
            for child_state, child_sign in spawn_children(st_end):
                queue.append((child_state, depth_n + 1, child_sign))

    return segments
```

```python
segments = generate_tree(depth_max=3, dt=0.005)

plt.figure(figsize=(7,7))
for seg in segments:
    plt.plot(seg[:,0], seg[:,1], lw=1)

plt.axis("equal")
plt.title("IG Fractal Tree (state inheritance + κ sign selection; ρ(s)=A^s)")
plt.show()
```
